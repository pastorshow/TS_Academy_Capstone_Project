**Early Warning Food Price Monitor — Northeast Nigeria**

### 1.2 Track and Problem Domain

This project was developed under the **Time Series Analysis
and Forecasting** track. It addresses the problem of food
price volatility in Northeast Nigeria — one of the most
food-insecure regions in Africa — by building a multi-model
forecasting pipeline capable of predicting monthly retail
food prices up to three months in advance.

Food price volatility in Northeast Nigeria is driven by a
compounding set of structural forces: the Boko Haram
insurgency, the June 2023 fuel subsidy removal, severe naira
devaluation, seasonal harvest cycles, and climate variability.
Existing price monitoring systems report prices only after
market changes have already occurred, providing no capacity
to anticipate future price movements. This project transforms
retrospective price monitoring into forward-looking market
intelligence.

## **Section 2 - Nigerian Food Security Context**

### 2.1 National Food Security Overview

Nigeria is home to over 220 million people, making it the most populous country in Africa, yet it remains one of the most food-insecure nations on the continent. According to the Food and Agriculture Organization of the United Nations (FAO) and the World Food Programme (WFP), approximately 25 million Nigerians faced acute food insecurity in 2023, a figure that has been rising steadily over the preceding decade. The Global Hunger Index (2023) classifies Nigeria's hunger level as serious, with the country ranking 109th out of 125 nations. The National Bureau of Statistics (NBS) reports that food inflation reached an all-time high of 37.92% in July 2024, eroding the purchasing power of millions of households who spend upwards of 60% of their income on food.

Nigeria's agricultural sector, which employs roughly 35% of the working-age population, is characterised by smallholder dominance, limited mechanisation, and heavy reliance on rain-fed production. The country produces significant quantities of staple commodities including rice, maize, yam, beans, and vegetables, yet remains structurally dependent on food imports — particularly for rice and wheat — to bridge domestic supply gaps. The Central Bank of Nigeria (CBN) reported that Nigeria's food import bill exceeded USD 5 billion annually in recent years, making the country highly exposed to global price shocks, exchange rate movements, and foreign exchange scarcity. This structural vulnerability means that disruptions to either domestic production or import supply chains translate rapidly into retail price spikes experienced by ordinary consumers.

### 2.2 The Northeast Nigeria Crisis: Adamawa, Borno, and Yobe

Among Nigeria's 36 states, the three northeastern states of Adamawa, Borno, and Yobe consistently register the highest rates of acute food insecurity in the country. The Cadre Harmonisé — the regional food security classification framework used across West Africa — routinely places large proportions of the population in these states in Crisis (Phase 3) or Emergency (Phase 4) conditions. In 2023, the Cadre Harmonisé estimated that over 4.3 million people across the Northeast faced acute food insecurity during the lean season, with Borno State accounting for the largest share of this caseload.

The primary driver of this crisis is the Boko Haram insurgency, which began in Borno State around 2009 and subsequently expanded into Adamawa and Yobe. Over more than a decade of conflict, the insurgency has devastated the agricultural foundations of the region. Farmers have been killed, displaced, or prevented from accessing their fields due to security threats, resulting in drastic reductions in cultivated land. The Lake Chad Basin — historically one of Nigeria's most productive agricultural zones — has seen a severe contraction in farming activity. The Armed Conflict Location and Event Data Project (ACLED) has documented thousands of conflict events in the tri-state region between 2015 and 2024, including attacks on rural markets, farmlands, and transportation routes.

The humanitarian consequences have been profound. The United Nations Office for the Coordination of Humanitarian Affairs (OCHA) estimated that the conflict had displaced over 2 million people within Nigeria by 2022, with the majority concentrated in Borno State. These internally displaced persons (IDPs) are predominantly farmers who have lost both their land and their livelihoods, creating a population almost entirely dependent on humanitarian food assistance. Market systems across the Northeast have been severely disrupted: supply chains are fragmented, traders face security risks on key transport corridors, and many rural markets have been abandoned or relocated. The combined effect is a food system that is structurally fragile, geographically constrained, and highly sensitive to further shocks.

### 2.3 Price Volatility and Its Drivers

Food price volatility in Northeast Nigeria is driven by an unusually dense concentration of overlapping structural and transitory factors, making the region one of the most challenging food market environments on the continent. Understanding these drivers is essential context for any price forecasting exercise.

Conflict remains the most immediate volatility driver. Market disruptions caused by insecurity reduce supply, increase transport costs, and limit trader activity, all of which cause localised price spikes that are often disconnected from national price trends. The removal of the fuel subsidy by the Tinubu administration in June 2023 constituted a second major structural shock: transport costs surged almost overnight, and since food distribution across Nigeria is overwhelmingly road-dependent, the cost increase was passed directly onto retail food prices. The World Bank estimated that the subsidy removal contributed to an acceleration in headline inflation, which crossed 30% by early 2024 according to NBS data.

Exchange rate depreciation has compounded these pressures. The naira lost more than 70% of its value against the US dollar between 2022 and 2024 following the liberalisation of the foreign exchange market, dramatically increasing the naira-denominated cost of imported rice, which is a key reference commodity in Nigerian food markets. Seasonal harvest cycles introduce a further layer of predictable volatility: prices typically peak during the lean season (May–August) before the main harvest, and fall after the September–November harvest period. However, in conflict-affected areas, this seasonal rhythm is disrupted by insecurity during planting and harvesting periods, creating unpredictable seasonal price dynamics. Rainfall anomalies — either flooding or drought — further compound supply uncertainty, particularly for rain-dependent crops such as yam and tomatoes.

### 2.4 The Case for Early Warning and Forecasting

Given this complex and compounding set of vulnerabilities, early warning systems are not merely useful but operationally critical. Humanitarian organisations such as WFP and NGOs operating in the Northeast must make procurement and pipeline decisions weeks or months in advance. Policymakers at the federal and state levels require lead time to activate strategic grain reserves, coordinate import licences, or deploy cash transfer programmes before a price crisis renders food inaccessible to the most vulnerable populations. Traders and market actors similarly benefit from price signals that allow them to make stocking and logistical decisions that can dampen rather than exacerbate price spikes.

A three-month forecasting horizon is specifically chosen in this project because it represents the minimum operational planning window for most humanitarian response actors while remaining within a temporal range where price dynamics are meaningfully predictable. Forecasting beyond six months in a conflict environment introduces too much uncertainty due to the non-stationarity of conflict drivers; forecasting only one month ahead provides insufficient lead time for logistical mobilisation.

### 2.5 Data Sources and Relevance

This project draws on five core data sources, each selected for its direct relevance to the food price dynamics described above. The WFP Vulnerability Analysis and Mapping (VAM) retail price dataset provides monthly retail prices for the five target commodities across markets in Adamawa, Borno, and Yobe, and constitutes the primary target variable for all forecasting models. CBN macroeconomic indicators — including exchange rates and money supply data — capture the monetary and trade dynamics that transmit global price shocks into domestic food markets. ACLED conflict event data provides a georeferenced record of security incidents that affect agricultural activity and market access in the Northeast. FAO import data tracks Nigeria's commodity import volumes and values, capturing supply-side disruptions driven by foreign exchange constraints. Finally, CHIRPS (Climate Hazards Group InfraRed Precipitation with Station data) rainfall data provides high-resolution precipitation records relevant to agricultural production cycles for rain-dependent commodities. Together, these sources provide a multi-dimensional picture of the drivers of food price variability in one of the world's most complex food crisis environments.

### 2.6 Connecting Context to This Project

The scale, complexity, and human cost of food insecurity in Northeast Nigeria make the development of robust, data-driven price forecasting tools both a technical and a humanitarian imperative. The Early Warning Food Price Monitor — Northeast Nigeria is designed to bridge the gap between existing price monitoring systems, which report historical trends, and the forward-looking intelligence that actionable humanitarian and policy responses demand. By integrating conflict, climate, macroeconomic, and market data into three complementary forecasting models — SARIMA, XGBoost, and Facebook Prophet — this project aims to provide a three-month price outlook for five critical commodities across three of Nigeria's most vulnerable states, giving decision-makers the lead time needed to act before crisis conditions become irreversible.

## **Section 3 — Problem Statement & Objectives**

### 3.1 Problem Statement
Food price volatility remains one of the most significant threats to food security and economic stability in Nigeria. Prices of staple commodities such as rice, beans, yam and tomatoes fluctuate sharply due to a combination of macroeconomic shocks, transportation costs, seasonal harvest cycles and conflict-related disruptions in agricultural regions. This volatility intensified following the removal of the petrol subsidy in June 2023, which created a structural break in Nigerian food markets and pushed prices to historically unprecedented levels. The consequences are severe for low-income households that spend a large proportion of their income on food, as well as for traders, farmers and policymakers who must make economic decisions under conditions of extreme uncertainty. Despite the scale of this challenge, existing food price monitoring systems in Nigeria are largely retrospective, reporting prices only after market changes have already occurred and providing little capacity to anticipate future price movements. As a result, government interventions such as food imports, subsidies and emergency assistance are typically implemented only after food price spikes have already harmed vulnerable populations. This project addresses this gap by developing a data-driven forecasting system capable of predicting monthly retail prices for key staple commodities across multiple Nigerian states up to three months in advance. By combining classical time series models with modern machine learning techniques, the system aims to provide forward-looking price intelligence that can support early warning systems, improve market planning and enable more proactive responses to food price instability.

### 3.2 Research Questions

RQ1: Which forecasting model — SARIMA, XGBoost or Prophet — produces the most accurate predictions of monthly retail food prices across Nigerian states?

RQ2: How does forecast accuracy change across different prediction horizons of one, two and three months ahead?

RQ3: Which commodity–state combinations produce sufficiently reliable forecasts to support operational deployment in a food price early warning system?

### 3.3 Project Objectives

To collect and preprocess monthly food price and economic indicator data covering January 2017 to December 2024 for five key commodities — Rice (imported), Rice (local), Beans (white), Yam and Tomatoes — across 13 Nigerian states, with modelling focused on the conflict-affected Northeast states of Adamawa, Borno and Yobe.
To integrate multiple drivers of food price movement, including macroeconomic indicators, rainfall patterns, fuel prices, import volumes and conflict events, into a unified modelling dataset.
To develop and train three forecasting models — SARIMA, XGBoost and Prophet — capable of generating one-month, two-month and three-month ahead retail food price predictions.
To evaluate and compare model performance against a naive baseline forecast using quantitative error metrics including Mean Absolute Percentage Error (MAPE), Mean Absolute Error (MAE) and Root Mean Squared Error (RMSE).
To generate policy-relevant insights and recommendations demonstrating how short-term food price forecasting can support early warning systems and improve decision-making for traders, farmers and government agencies.


### 3.4 Project Significance
Food price instability directly threatens the welfare of millions of Nigerian households who depend on affordable staple foods for daily consumption. A reliable forecasting system would enable policymakers and humanitarian organisations to anticipate price spikes and implement interventions before food insecurity worsens. By transforming historical price monitoring into forward-looking market intelligence, this project demonstrates how data-driven forecasting can strengthen Nigeria's capacity to respond proactively to food price shocks. In the long term, such systems could contribute to more stable food markets and improved resilience for vulnerable populations.





## **Section 4 — Literature Review**

### Purpose

This section reviews key literature on food price forecasting, time series methods, and Nigerian agricultural food security studies to provide a theoretical foundation for this project's approach to forecasting food prices in Northeast Nigeria. It highlights developments in forecasting methodology, the demonstrated strengths and limitations of SARIMA, XGBoost, and Prophet in agricultural price contexts, and the specific structural conditions — conflict, macroeconomic shocks, and seasonal volatility — that make the Northeast Nigeria forecasting problem both challenging and consequential.

---

### 4.1 Food Price Forecasting

Food price forecasting has evolved considerably over the past two decades, moving from traditional econometric models toward hybrid machine learning approaches driven by the need to address volatility in agricultural commodities influenced by supply chain disruptions, climate variability, and geopolitical conflict. Early studies emphasised statistical time series methods for predicting price indices of staple commodities such as cereals, millets, and pulses, demonstrating that accurate price forecasts can reduce food wastage and improve supply chain management [10]. Research on wholesale food prices in European markets showed that retail and wholesale prices exhibit sufficient temporal structure to support automated forecasting, and that time series methods could meaningfully reduce the burden on manual pricing decisions [9].

More recent literature has expanded the scope of food price forecasting to include exogenous variables such as macroeconomic indicators, exchange rate movements, commodity futures prices, and conflict indices. Studies show that univariate models frequently underperform in volatile markets where external shocks — rather than historical price patterns alone — drive price behaviour [1][4]. Research on adaptive forecasting during periods of rapid economic change, including assessments of the U.S. Food Price Outlook, demonstrates that incorporating uncertainty signals and external economic indicators meaningfully improves forecast accuracy over benchmark approaches [5][7].

In developing country contexts, food price forecasting serves a direct humanitarian function. Price spikes translate immediately into reduced caloric intake for low-income households, making forward-looking price intelligence a tool for preventive action rather than retrospective analysis [3][6]. The literature consistently identifies a gap in applications to conflict-affected environments, where the standard assumptions of market integration and supply chain continuity break down, and where volatility is amplified by non-economic drivers such as insecurity and displacement. This project directly addresses that gap.

---

### 4.2 Time Series Forecasting Methods

Three modelling approaches — SARIMA, XGBoost, and Facebook Prophet — were selected for this project based on their demonstrated performance in food price and agricultural forecasting literature, and their complementary methodological strengths.

#### SARIMA

SARIMA — Seasonal AutoRegressive Integrated Moving Average — extends the classical ARIMA framework with an explicit seasonal component, making it well-suited to monthly agricultural price series that exhibit recurring annual cycles. It has been widely applied to food commodity price indices and has demonstrated reliable performance in stable, low-volatility settings where historical price patterns carry strong predictive signal [10][13]. Comparative studies consistently position SARIMA as a robust statistical baseline, one that is frequently competitive with machine learning models when price behaviour is regular and seasonal structure is dominant [9][29]. Its principal limitation is sensitivity to structural breaks — abrupt changes in price level or dynamics caused by external shocks that lie outside its training distribution. In these settings, SARIMA's reliance on historical autocorrelation structure produces systematic forecast errors, and machine learning or decomposition-based approaches tend to outperform it [1][12].

#### XGBoost

XGBoost — Extreme Gradient Boosting — is a tree-based ensemble learning algorithm capable of capturing complex non-linear relationships between multiple features and a target variable. Its ability to incorporate exogenous variables such as conflict indicators, exchange rates, fuel prices, and rainfall data alongside price history makes it a natural candidate for food price forecasting in environments where multiple structural drivers operate simultaneously [7][11]. In demand and sales forecasting applications, XGBoost has demonstrated competitive or superior performance relative to SARIMA on datasets with strong feature-target relationships and sufficient training volume [12][23]. However, a well-documented limitation of tree-based models is their inability to extrapolate reliably beyond the range of values seen during training. In post-shock environments where prices rise sharply to levels never observed in the historical training data, XGBoost predictions are constrained by their training distribution, producing systematic under-prediction [8]. This limitation is directly relevant to the post-subsidy price environment in this project.

#### Facebook Prophet

Prophet is a decomposable additive time series model developed by Meta Research that represents price as the sum of trend, seasonality, and external regressor components. It was designed to handle irregular time series with missing data, multiple seasonality layers, and known structural breaks, and has demonstrated strong performance in agricultural and food price applications where trend and seasonality co-exist [9][19]. Comparative analyses show Prophet outperforming SARIMA in computation efficiency and accuracy for food price indices, particularly in settings with a clear long-run trend [9][15]. Its trend component allows it to project price trajectories into territory beyond the training range — a meaningful advantage in post-shock forecasting environments where prices accelerate past historically observed levels. Hybrid approaches combining Prophet with ARIMA or machine learning components have been proposed for edge environments requiring both linear trend projection and non-linear feature interaction [19], though for the scale and data availability of this project, the standalone Prophet configuration with an external regressor for the structural break was adopted.

#### Comparative Evidence

Comparative studies that evaluate all three model families on time series data with seasonal structure and exogenous drivers generally find that model performance is highly context-dependent [13][29]. SARIMA performs well under stability; XGBoost performs well when feature-target relationships are strong and the test period falls within the training distribution; Prophet performs well when trend and seasonality are the dominant price-generating mechanisms. No single model universally dominates across all commodity-state environments — a finding consistent with this project's results, where Prophet won 10 of 13 Tier 1 combinations, XGBoost won 2, and SARIMA won 1.

### 4.3 Nigerian Agriculture and Food Security

Nigerian agriculture, particularly in the Northeast, faces a compounding set of structural challenges that directly shape food price dynamics across the 2017–2024 observation window of this project. Understanding these conditions is not merely contextual — they determine the difficulty of the forecasting task and the appropriate interpretation of model results.

The Boko Haram insurgency, which began in Borno State around 2009 and subsequently spread into Adamawa and Yobe, has fundamentally altered the agricultural economy of the Northeast. Research documents a 47% reduction in cultivated land across the Borno-Adamawa-Yobe (BAY) states as a direct consequence of conflict-driven displacement, farm abandonment, and restricted access to agricultural land [36][37]. The Lake Chad Basin — historically one of Nigeria's most productive farming zones — has experienced a severe contraction in agricultural output, disrupting the regional supply chains that underpin food availability across the three states [37][38]. The Armed Conflict Location and Event Data Project (ACLED) records thousands of conflict events in the tri-state region between 2015 and 2024, including direct attacks on rural markets, farmland, and transportation routes that are critical to food distribution.

The humanitarian consequences are severe and persistent. The Cadre Harmonisé, the regional food security classification framework used across West Africa, has consistently placed large proportions of the BAY state population in Crisis (Phase 3) or Emergency (Phase 4) conditions throughout the project's observation window [31][32]. WFP assessments reported 4.8 million people across the BAY states facing acute food insecurity in 2024, with 57.7% of households reporting inadequate food consumption [33]. Malnutrition rates have surged alongside food insecurity, with severe acute malnutrition admissions running 70% higher in early 2024 compared to the same period in 2023 [32][34]. The United Nations Office for the Coordination of Humanitarian Affairs estimated that conflict had displaced over 2 million people within Nigeria by 2022, the majority concentrated in Borno State, creating a population almost entirely dependent on humanitarian food assistance [32].

Food price volatility in the Northeast is driven by an unusually dense concentration of overlapping structural and transitory factors. Conflict reduces market supply, increases transport costs, and limits trader activity — producing localised price spikes that frequently diverge from national price trends [36][38]. The removal of the fuel subsidy by the Tinubu administration in June 2023 constituted a second major structural shock: transport costs increased sharply overnight, and since food distribution across Nigeria is overwhelmingly road-dependent, this cost increase translated directly and rapidly into retail food prices [39]. The naira's loss of more than 70% of its value against the US dollar between 2022 and 2024, following foreign exchange market liberalisation, further elevated the naira-denominated cost of imported rice — a key reference commodity in Nigerian food markets. Seasonal harvest cycles introduce a further layer of predictable volatility, though in conflict-affected areas this seasonal rhythm is disrupted by insecurity during planting and harvesting periods, creating irregular seasonal dynamics that are difficult for standard time series models to capture [37].

Despite the scale and complexity of this food security crisis, the application of rigorous time series forecasting methods to Nigerian Northeast food prices remains limited in the academic literature. Existing work on Nigerian food price dynamics has focused primarily on national-level trends or macroeconomic drivers, without integrating conflict event data, walk-forward multi-horizon evaluation, or the kind of commodity-state level granularity that operational decision-making requires [36][40]. This project addresses that gap directly by combining SARIMA, XGBoost, and Prophet with conflict, climate, and macroeconomic features at the commodity-state level, evaluated across three forecast horizons in one of the world's most complex food crisis environments.

#### Economic Context — Four Regime Periods

The 2017–2024 observation window spans four
distinct economic regimes that significantly shaped
Nigerian food price behaviour.

| Period | Dates | Key Characteristic |
|--------|-------|--------------------|
| Pre-COVID Baseline | Jan 2017 – Feb 2020 | Stable naira, moderate inflation, consistent supply chains |
| COVID Disruption | Mar 2020 – Dec 2021 | Border closures, supply chain shocks, demand contraction |
| Post-COVID Recovery | Jan 2022 – May 2023 | Gradual price normalisation, naira pressure building |
| Post-Subsidy Shock | Jun 2023 – Dec 2024 | Fuel subsidy removal, severe naira devaluation, rapid price acceleration |

### 5.3 Feature Engineering

#### Overview

Feature engineering was conducted in two phases.
The first phase occurred in Notebook 03 during
dataset merging, producing three engineered features
that were added to the master dataset and used by
all downstream models. The second phase occurred
within each modelling notebook, where model-specific
transformations were applied immediately before
training.

---

#### Phase 1 — Master Dataset Features (Notebook 03)

Three features were engineered and saved permanently
into the master dataset.

---

**Feature 1 — Price Lag Features**

Food prices exhibit strong autocorrelation — this
month's price is highly predictable from last month's
price. EDA confirmed this with correlation values of
r = 0.968 for price_lag1, r = 0.953 for price_lag2,
and r = 0.924 for price_lag3 against the target
variable. These are the three strongest predictors
in the entire feature set.

Three lag features were created per commodity-state
combination:

| Feature | Definition |
|---------|-----------|
| price_lag1 | Retail price 1 month prior |
| price_lag2 | Retail price 2 months prior |
| price_lag3 | Retail price 3 months prior |

Implementation: The dataset was sorted by state,
commodity and date before applying pandas shift()
within each commodity-state group. Sorting before
shifting is critical — unsorted data would assign
lags from the wrong time period, introducing
data leakage.

The first 1, 2 and 3 months of each commodity-state
series produce null values representing the warmup
period where insufficient history exists. These nulls
(71 for lag1, 141 for lag2, 211 for lag3) were
resolved in each modelling notebook using forward
fill and backward fill within groups immediately
before model training.

---

**Feature 2 — Harvest Season Indicator**

Nigerian agricultural commodities follow predictable
annual harvest cycles. During harvest months, supply
increases and retail prices typically fall. A binary
harvest_season indicator tells each model whether
a given observation falls in a harvest month for
that commodity.

Harvest month definitions were set according to the
Nigerian agricultural calendar:

| Commodity | Harvest Months |
|-----------|---------------|
| Maize (white) | August, September, October |
| Maize (yellow) | August, September, October |
| Rice (imported) | October, November, December |
| Rice (local) | October, November, December |
| Beans (white) | October, November |
| Yam | August, September, October |
| Tomatoes | November, December, January |

Implementation: A dictionary mapping each commodity
to its harvest months was defined. A row-wise lambda
function checks whether the month of each row falls
within the harvest months for that row's commodity
and assigns 1 (harvest) or 0 (non-harvest).

Result: 727 harvest month rows (25.5%) and 2,119
non-harvest rows (74.5%) — consistent with a cycle
of 2–3 harvest months per commodity across a
12-month year.

Note: EDA revealed that harvest_season showed
weaker-than-expected correlation with retail prices.
Market integration and storage dynamics meant that
retail prices do not always fall immediately during
harvest months. This was documented in EDA Step 8
and the feature was retained as a directional signal
rather than a strong predictor.

---

**Feature 3 — Other Commodity Average Price**

Food commodity prices influence each other through
substitution and complementarity effects. When one
staple becomes expensive, consumers shift demand
toward cheaper alternatives, putting upward pressure
on substitute commodities. Farmers also make
planting decisions based on relative commodity
profitability.

This feature captures aggregate cross-commodity
price pressure in each state and month as a single
summary signal.

Implementation: The mean price across all commodities
was computed for each state-month combination using
groupby mean. This aggregate was then merged back
onto the master dataset on date and state. Each row
receives the average price of all other commodities
in its state and month.

---

#### Phase 2 — Model-Specific Transformations
(Applied in Modelling Notebooks)

Several additional transformations were applied
within individual modelling notebooks rather than
at the master dataset level. This was deliberate —
these transformations are model-specific and applying
them to the master dataset would have coupled
modelling decisions to the shared data layer.

---

**Log transformation of target variable (NB05, NB06, NB07)**

The target variable price_ngn was log-transformed
using log1p (natural log of price + 1) before
model training in all three modelling notebooks.
```python
df['log_price'] = np.log1p(df['price_ngn'])
```

Reason: Food prices are right-skewed with occasional
extreme spikes. Log transformation compresses the
scale, stabilises variance, and converts multiplicative
price growth into additive increments — a more
appropriate representation for regression models.
All predictions were inverse-transformed back to
Naira scale using expm1 before metric calculation
to ensure MAPE and RMSE are interpretable.

---

**Subsidy removal indicator (NB05, NB06, NB07)**

A binary structural break indicator was engineered
to explicitly mark the June 2023 fuel subsidy removal
— the single most consequential event in the dataset.
```python
df['subsidy_removed'] = (df['date'] >= '2023-06-01').astype(int)
```

The value is 0 for all observations before June 2023
and 1 from June 2023 onward. This feature gives each
model a direct signal that price behaviour changed
structurally at this point. In Prophet it was used
as an external regressor. In XGBoost it was included
in the feature set.

Note: A single known changepoint date of June 2023
was initially considered for Prophet's built-in
add_changepoints parameter but was rejected because
Prophet rejects changepoint dates that fall outside
the training window boundary. The external regressor
approach was used instead.

---

**Log transformation of import volume (NB06)**

Import volume (import_volume_tonnes) was
log-transformed before use in XGBoost to correct
its heavily right-skewed distribution identified
in EDA.
```python
df['log_import_volume'] = np.log1p(
    df['import_volume_tonnes'].clip(lower=0)
)
```

The clip(lower=0) guard prevents negative values
from causing log transformation errors.

---

**Categorical encoding of state and commodity (NB06)**

XGBoost requires numerical inputs. State and commodity
were label-encoded using sklearn's LabelEncoder.
```python
df['state_encoded']     = LabelEncoder().fit_transform(df['state'])
df['commodity_encoded'] = LabelEncoder().fit_transform(df['commodity'])
```

Label encoding was chosen over one-hot encoding
to keep the feature space compact given the small
per-combination training sizes of 60–90 observations.

---

#### Feature Exclusions and Rationale

Two features were collected and merged into the
master dataset but explicitly excluded from all
modelling:

| Feature | Correlation | Retained Feature | Reason for Exclusion |
|---------|-------------|-----------------|---------------------|
| conflict_events | r = 0.99 with conflict_score_weighted | conflict_score_weighted | Near-identical information. Retaining both splits feature importance without adding signal |
| mpr_pct | r = 0.88 with pms_price_ngn | pms_price_ngn | High redundancy with fuel price. Retaining both dilutes the fuel price importance signal |

---

#### Final Feature Set Summary

| Feature | Engineering Phase | Type |
|---------|------------------|------|
| price_lag1, price_lag2, price_lag3 | NB03 | Lag features |
| harvest_season | NB03 | Calendar indicator |
| other_commodity_avg_price | NB03 | Cross-commodity signal |

### 5.4 Model Architecture and Implementation

#### Overview

Three forecasting models were built and evaluated:
SARIMA (Notebook 05), XGBoost (Notebook 06), and
Facebook Prophet (Notebook 07). A naive baseline
was established identically in all three notebooks
to ensure all comparisons are made on equal terms.
Each model was evaluated on the same Tier 1
commodity-state combinations covering Adamawa,
Borno and Yobe. Prophet was additionally extended
to Tier 2 states.

---

#### Naive Baseline

The naive baseline predicts that next month's price
equals the last known price before the test period.
It requires no training and makes no assumptions
about structure. It serves as the minimum performance
floor — any model that cannot outperform it has
no practical value.

The same naive baseline logic was implemented
identically in NB05, NB06 and NB07 to guarantee
fair comparison across models.

---

#### Model 1 — SARIMA (Notebook 05)

SARIMA — Seasonal AutoRegressive Integrated Moving
Average — is the classical statistical baseline
for time series forecasting. It models price as
a function of its own past values and past
forecast errors, with an explicit seasonal
component.

SARIMA parameters were determined individually
per commodity-state combination rather than using
a single global parameter set. This was done
because different commodities exhibit different
autocorrelation structures and seasonal patterns.

SARIMAX — the exogenous variable extension —
was also tested across three feature configurations:
macroeconomic features only, conflict features
added, and all features combined. In all three
configurations SARIMAX failed to consistently
outperform the baseline SARIMA, and SARIMA
was retained as the statistical model.

SARIMA was evaluated on Tier 1 only (13
commodity-state combinations) using single-step
forecasting. Multi-horizon evaluation was not
implemented in this notebook — this is a known
limitation documented in the model comparison.

---

#### Model 2 — XGBoost (Notebook 06)

XGBoost — Extreme Gradient Boosting — is a
tree-based ensemble model that learns non-linear
relationships between features and the target.
Unlike SARIMA it can simultaneously incorporate
all engineered features alongside price history.

**Direct multi-horizon forecasting:**
Rather than using recursive forecasting (feeding
predictions back as inputs), three separate
XGBoost models were trained per commodity-state
combination — one each for the 1-month, 2-month
and 3-month ahead horizons. This direct approach
avoids compounding prediction errors that occur
in recursive strategies.

**Feature set used:**
price_lag1, price_lag2, price_lag3,
harvest_season, other_commodity_avg_price,
log_import_volume, conflict_score_weighted,
conflict_fatalities, exchange_rate_ngn_usd,
inflation_rate_pct, pms_price_ngn,
rainfall_mm, subsidy_removed,
state_encoded, commodity_encoded

**Target variable:** log_price (log1p transformed),
inverse-transformed before metric calculation.

XGBoost was evaluated on Tier 1 only
(13 combinations × 3 horizons = 39 models).

---

#### Model 3 — Facebook Prophet (Notebook 07)

Prophet is a decomposable time series model
developed by Meta that models price as the
sum of trend, seasonality and external regressor
components. It is designed to handle structural
breaks and missing data robustly.

**Key configuration decisions:**

| Setting | Choice | Reason |
|---------|--------|--------|
| Seasonality mode | Multiplicative | Food prices scale proportionally — a 20% seasonal effect at ₦500 is different from 20% at ₦2000 |
| Yearly seasonality | Enabled | All commodities show annual cycles confirmed in EDA |
| External regressor | subsidy_removed | Captures June 2023 structural break as a known binary shift |
| Changepoints | Not manually set | Prophet rejected a June 2023 changepoint date falling outside the training window boundary |

**Walk-forward evaluation:**
Prophet implemented walk-forward evaluation at
1, 2 and 3-month horizons — an improvement over
SARIMA's single-step evaluation that enables
direct horizon-by-horizon comparison with XGBoost.

**Tier 1 results:** 13/13 combinations fitted
successfully. Prophet outperformed the naive
baseline on all 13 combinations.

**Tier 2 extension:**
Prophet was extended to all 10 Tier 2 states
(44 commodity-state combinations). Because fixed
date split boundaries produced zero test
observations for some Tier 2 combinations with
shorter coverage, a proportional split function
(get_split_tier2) was implemented that allocates
a fixed percentage of each series to training
and testing regardless of its absolute length.

A credibility filter was applied to Tier 2 results:
combinations with average MAPE above 35% were
classified as directional-only rather than
production-ready. 24 of 44 Tier 2 combinations
passed the credibility filter.



### 5.5 Evaluation Framework

#### Primary Metric — MAPE

Mean Absolute Percentage Error was used as the
primary evaluation metric across all models and
all comparisons.

MAPE = (1/n) × Σ |actual - predicted| / actual × 100

MAPE was chosen because it is scale-independent —
it produces comparable error figures across
commodities with very different price levels.
A 20% MAPE for Yam (higher absolute price) and
a 20% MAPE for Beans (lower absolute price)
represent equivalent forecasting accuracy.
This makes cross-commodity comparison meaningful.

All MAPE calculations were performed on the
original Naira scale after inverse-transforming
log predictions back to prices using expm1.
This is important — computing MAPE on log-scale
predictions would produce artificially deflated
error figures.

---

#### Secondary Metrics — MAE and RMSE

**MAE (Mean Absolute Error)** measures average
absolute prediction error in Naira. It is
directly interpretable — an MAE of ₦200 means
predictions are on average ₦200 from the
actual price.

**RMSE (Root Mean Squared Error)** penalises
large errors more heavily. It is sensitive to
spikes and regime shifts, making it a useful
complement to MAPE for identifying whether
errors are uniformly distributed or concentrated
in high-volatility periods.

---

#### Multi-Horizon Evaluation

Forecasts were evaluated at three separate
horizons:

| Horizon | Definition |
|---------|-----------|
| H1 | 1-month ahead forecast |
| H2 | 2-month ahead forecast |
| H3 | 3-month ahead forecast |

Evaluating across horizons is critical for
assessing whether model accuracy degrades
predictably as the forecast window extends.
Healthy degradation (H1 < H2 < H3 in MAPE)
signals the model is capturing genuine
uncertainty rather than overfitting to the
test period.

SARIMA implemented single-step forecasting only
(H1 equivalent). Its results were reported under
H1_MAPE with H2 and H3 set to null in the model
comparison notebook to avoid misleading
aggregations.

---

#### Baseline Comparison

All model results were compared against the
naive baseline MAPE for each combination.
A model was only considered to add value if
its MAPE was meaningfully lower than the naive
baseline — not marginally.

---

#### Tier 2 Credibility Filter

A MAPE threshold of 35% was applied to Tier 2
Prophet results. Combinations above this threshold
were labelled directional-only, meaning they may
capture the direction of price movement but
cannot be relied upon for precise forecasting.
This distinction was preserved in the results
CSV output for downstream use.

---

#### Model Comparison — Notebook 08

All results were aggregated in Notebook 08 using
three CSV files output from the modelling notebooks:
sarima_results.csv, prophet_results.csv and
xgb_results.csv. The best model per commodity-state
combination was determined by lowest average MAPE
across available horizons. Results were saved as
model_comparison_final.csv and
best_model_per_combination.csv.


### 5.6 Train, Validation and Test Split Design

#### Core Principle

A temporal split was used throughout the project.
Random splitting was explicitly rejected for all
models. In time series forecasting, random splitting
breaks the temporal order of observations, allowing
the model to train on future data and evaluate on
past data — producing optimistic but invalid results.
Every split in this project respects chronological
order: the model never sees any data point that
occurs after the data it is being evaluated on.

---

#### Tier 1 Split — Fixed Date Boundaries

For Tier 1 states (Adamawa, Borno, Yobe) fixed date
boundaries were applied uniformly across all
commodity-state combinations:

| Split | Period | Months |
|-------|--------|--------|
| Training | January 2017 – December 2022 | 72 months |
| Validation | January 2023 – May 2023 | 5 months |
| Test | June 2023 – December 2024 | 19 months |

The training set covers the full pre-subsidy
history. The validation set covers the transition
period before the subsidy removal. The test set
begins at June 2023 — deliberately coinciding with
the fuel subsidy removal — to ensure evaluation
captures the most challenging and policy-relevant
period in the dataset.

The validation set was used for SARIMAX feature
selection in Notebook 05. In Prophet and XGBoost,
walk-forward evaluation was performed on the test
set directly.

---

#### Tier 2 Split — Proportional Boundaries

Tier 2 states have shorter and less uniform data
coverage than Tier 1. Applying the same fixed
date boundaries to Tier 2 combinations produced
zero test observations for multiple combinations
where data ended before June 2023.

A proportional split function (get_split_tier2)
was implemented to resolve this:

- Training: first 70% of each series chronologically
- Validation: next 15%
- Test: final 15%

This guarantees every Tier 2 combination has a
non-empty test set regardless of its absolute
length. The make_future_dataframe call in Prophet
was explicitly extended to cover the full
proportional test period rather than using the
default 30-day horizon.

---

#### Why the Test Period Starts at June 2023

The test set begins at the subsidy removal date
by design. This was a deliberate methodological
decision, not an arbitrary date choice. The period
June 2023 to December 2024 is the hardest
forecasting period in the dataset — prices
accelerated sharply, the naira devalued, and
fuel costs spiked simultaneously. A model that
performs well on this period demonstrates genuine
robustness to structural shocks, not just
accuracy under stable conditions.

Post-subsidy systematic under-prediction observed
in all three models was documented and correctly
interpreted as distributional shift — prices in
the test period exceeded any level seen in the
training data. This is not model misspecification;
it reflects the limits of any model trained on
pre-shock data being evaluated on a post-shock
regime.

A core question in any forecasting system is how accuracy degrades as predictions are made further into the future. This project evaluated all models at three horizons — one month ahead (H1), two months ahead (H2), and three months ahead (H3). SARIMA was evaluated at H1 only due to its single-step architecture. Prophet and XGBoost were evaluated across all three horizons.

| Horizon | Prophet Avg MAPE | XGBoost Avg MAPE |
|:---|:---:|:---:|

| H3 — 3 months ahead | 35.3% | 49.7% |

Prophet shows a healthy and expected pattern of horizon decay — accuracy falls gradually from 24.5% at H1 to 35.3% at H3. XGBoost shows an unusual pattern — its accuracy does not decay meaningfully across horizons and remains consistently weak at all three. This is not a sign of stability but rather of a model that was already struggling at H1 due to the structural break introduced by the 2023 fuel subsidy removal, and continued to struggle equally at H2 and H3 for the same reason.

### 6.5 The Impact of the 2023 Fuel Subsidy Removal

All three models were trained on price data from before June 2023. When the Nigerian government removed the fuel subsidy in June 2023, food prices jumped sharply to levels the models had never encountered in their training data. The result was a structural break — a moment where the historical patterns the models had learned were no longer a reliable guide to what prices would do next.

This is like asking someone to predict the price of petrol using only data from before a price cap was removed. Once the cap lifts and prices leap to a new level, predictions based on the old pattern will systematically underestimate the new reality — not because the model is poorly built, but because the world it was trained on no longer exists.

Prophet handled this disruption better than XGBoost for a specific reason: it is designed to project a trend line forward even into price territory it has never seen before. When prices surged past the training range, Prophet extended its trend estimate forward and partially tracked the new levels. XGBoost struggles to reliably predict values significantly higher than those seen during training because it has no mechanism for extrapolation beyond its training range. This is a key reason why XGBoost's accuracy was consistently weak across all three forecast horizons — from the very first month of the test period, it was already operating outside its comfort zone, and this structural limitation persisted regardless of how far ahead it was asked to predict.

The subsidy removal was the single biggest external factor affecting these results. It did not invalidate the models — all 13 combinations remained usable despite the shock — but it does underline the need to retrain models regularly as new price data becomes available, particularly following major policy changes.

### 6.6 Key Takeaways

- All 13 Tier 1 combinations produced usable forecasts — 11 production-ready, 2 directional only
- Prophet is the recommended model for 10 of 13 combinations, including all three Excellent-rated results
- The best result in the project: Rice (imported) in Borno — Prophet — 14.7% MAPE
- XGBoost is preferred for Yam in Adamawa (21.7%) and Rice (local) in Adamawa (30.9%) — the only state where Prophet was not dominant
- SARIMA is competitive only on Beans (white) in Borno (25.6%), where price stability means historical patterns alone carry enough signal
- Prophet's accuracy decays gradually and healthily across horizons: 24.5% at H1, 28.5% at H2, 35.3% at H3 — all within acceptable ranges
- The 2023 fuel subsidy removal remains the defining challenge for all models — regular retraining is essential to maintain accuracy going forward

## **Section 7 — Conclusion, Limitations & References**

### 7.1 Conclusion

This project set out to answer a straightforward but consequential question: can monthly retail food prices in Northeast Nigeria be forecast accurately enough, and far enough in advance, to support real humanitarian and policy decisions? The answer, based on the results of this evaluation, is yes — with important qualifications.

Across 13 commodity-state combinations in Adamawa, Borno and Yobe, three forecasting models were built, trained and evaluated against a naive baseline across one-month, two-month and three-month forecast horizons. The central finding is that 11 of 13 combinations achieved production-ready accuracy, meaning forecast errors were low enough to support operational planning decisions. The remaining 2 combinations produced directional forecasts — useful for understanding the direction of price movement, though not precise enough for exact planning.

Prophet was the dominant model, winning 10 of 13 combinations with an average MAPE of 30.8% across all horizons. Its ability to decompose Nigerian food price series into a long-run trend and an annual seasonal cycle — and to project that trend into post-shock territory — made it the most appropriate model for a price environment shaped simultaneously by structural inflation, naira devaluation and the June 2023 fuel subsidy removal. XGBoost won 2 combinations, performing best where short-term economic signals carried more predictive power than trend decomposition. SARIMA won 1 combination, demonstrating that classical statistical models remain competitive on stable, seasonally regular commodities.

The single best result in the evaluation was Prophet on Rice (imported) in Borno State — a MAPE of 14.7%, meaning the model's predictions were off by less than 15% of the actual price on average across the full 19-month test period. This is a strong result by any forecasting standard, achieved despite the model being evaluated entirely on post-subsidy data it had never seen during training.

The Early Warning Food Price Monitor — a live web application built from these results — demonstrates that the forecasting pipeline can be operationalised. Decision-makers in Adamawa, Borno and Yobe can access three-month price outlooks for five key commodities without requiring any technical knowledge. This moves the project from academic exercise to deployable early warning tool.

---

### 7.2 Limitations

Honest acknowledgement of limitations is essential for any forecasting system intended for real-world use. Four limitations are documented here.

**Structural break sensitivity.** All three models were trained on data from before June 2023. The fuel subsidy removal in June 2023 created a structural break — a sudden shift in price levels that the models had no training basis for. The consequence was systematic under-prediction in the test period, particularly in the first months after the shock. This is not model failure; it is a known limitation of any model trained on pre-shock data and evaluated on post-shock conditions. The implication is that models must be retrained regularly, especially following major policy changes or economic shocks, to maintain accuracy.

**SARIMA single-horizon limitation.** SARIMA was evaluated at the one-month ahead horizon only. It does not natively support multi-step direct forecasting in the way XGBoost and Prophet do. This means SARIMA's performance figures are not directly comparable to those of the other two models at H2 and H3, and its competitiveness may be understated or overstated depending on how it would perform at longer horizons.

**Tier 2 coverage.** Of 44 Tier 2 commodity-state combinations evaluated using Prophet, only 24 passed the production-ready credibility threshold. The remaining 20 combinations produced directional-only results. Tier 2 states have shorter and less consistent price histories than the core Tier 1 states, and this data limitation directly constrains forecast reliability. The forecasting system in its current form is most reliable for Adamawa, Borno and Yobe.

**External feature availability.** The modelling pipeline incorporates macroeconomic, conflict, rainfall and import volume data. In a live deployment context, some of these features — particularly ACLED conflict data and CBN macroeconomic indicators — may not be available in time for a given month's forecast. A real-time deployment would need to address the lag between data generation and data availability for each external feature.

---

### 7.3 Recommendations

**Retrain models quarterly.** The most important operational recommendation is to retrain all models as new price data becomes available, at minimum every quarter. The post-subsidy results demonstrate clearly that models degrade when price regimes shift. A model trained on data through 2024 will be significantly more accurate in 2025 than the models evaluated in this project.

**Prioritise Prophet for deployment.** For the 10 combinations where Prophet was the winning model, it should be the primary forecasting engine in any operational deployment. Its trend projection capability makes it the most robust choice in an environment where prices continue to be shaped by structural forces — inflation, exchange rate pressure, and fuel costs — rather than short-term fluctuations alone.

**Extend the system to Tier 2 states.** With more consistent data collection and a longer price history for Tier 2 states, the forecasting system could be extended reliably to cover all 13 states in the dataset. This would significantly increase the humanitarian reach of the early warning system.

**Integrate the system with existing early warning platforms.** The Early Warning Food Price Monitor should be positioned as a complement to existing tools such as the WFP VAM platform and the Cadre Harmonisé. Integration with these systems — sharing forecast outputs as inputs to broader food security assessments — would maximise the operational value of the three-month price outlook.

**Explore ensemble approaches in future work.** A model averaging or stacking approach that combines Prophet and XGBoost predictions could reduce the variance of forecast errors, particularly for high-volatility commodities like Yam and Tomatoes where no single model is consistently dominant. This is the most promising direction for future technical development.

---

### 7.4 Final Remarks

Food price forecasting in conflict-affected environments is one of the harder applied problems in data-driven humanitarian work. The combination of structural shocks, seasonal volatility, conflict-driven disruptions and data scarcity means that no model will ever forecast perfectly. What this project demonstrates is that rigorous, multi-model forecasting with honest evaluation produces results that are good enough to be genuinely useful — and that useful is what matters when the alternative is no forward-looking price intelligence at all.

Three months of advance warning is not a technical output. It is time — time for a policymaker to activate a reserve, for an aid organisation to adjust a procurement pipeline, for a trader to make a stocking decision that keeps food accessible rather than scarce. This project exists to create that time.
